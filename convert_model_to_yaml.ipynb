### 3. 함수를 통해 model_info_txt에서 model_config['backbone']을 뽑아낸다

```python
def search_model():
    """Search model structure from user-specified search space."""
    model = []
    n_stride = 0
    MAX_NUM_STRIDE = 5
    UPPER_STRIDE = 2  # 5(224 example): 224, 112, 56, 28, 14, 7

    # Module 1
    m1 = model_info_txt.params['m1']
    m1_args = []
    m1_repeat = model_info_txt.params['m1/repeat']
    m1_out_channel = model_info_txt.params['m1/out_channels']
    m1_stride = model_info_txt.params['m1/stride']
    if m1_stride == 2:
        n_stride += 1
    m1_activation = model_info_txt.params['m1/activation']
    if m1 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m1_args = [m1_out_channel, 3, m1_stride, None, 1, m1_activation]
    elif m1 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m1_args = [m1_out_channel, 3, m1_stride, None, m1_activation]
    model.append([m1_repeat, m1, m1_args])

    # Module 2
    m2 = model_info_txt.params['m2']
    m2_args = []
    m2_repeat = model_info_txt.params['m2/repeat']
    m2_out_channel = model_info_txt.params['m2/out_channels']
    m2_stride = model_info_txt.params['m2/stride']
    # force stride m2
    if n_stride == 0:
        m2_stride = 2
    if m2 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m2_kernel = model_info_txt.params["m2/kernel_size"]
        m2_activation = model_info_txt.params["m2/activation"]
        m2_args = [m2_out_channel, m2_kernel, m2_stride, None, 1, m2_activation]
    elif m2 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m2_kernel = model_info_txt.params["m2/kernel_size"]
        m2_activation = model_info_txt.params["m2/activation"]
        m2_args = [m2_out_channel, m2_kernel, m2_stride, None, m2_activation]
    elif m2 == "InvertedResidualv2":
        m2_c = model_info_txt.params["m2/v2_c"]
        m2_t = model_info_txt.params["m2/v2_t"]
        m2_args = [m2_c, m2_t, m2_stride]
    elif m2 == "InvertedResidualv3":
        m2_kernel = model_info_txt.params["m2/kernel_size"]
        m2_t = model_info_txt.params["m2/v3_t"]
        m2_c = model_info_txt.params["m2/v3_c"]
        m2_se = model_info_txt.params["m2/v3_se"]
        m2_hs = model_info_txt.params["m2/v3_hs"]
        # k t c SE HS s
        m2_args = [m2_kernel, m2_t, m2_c, m2_se, m2_hs, m2_stride]
    if not m2 == "Pass":
        if m2_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m2_repeat, m2, m2_args])

    # Module 3
    m3 = model_info_txt.params["m3"]
    m3_args = []
    m3_repeat = model_info_txt.params["m3/repeat"]
    m3_stride = model_info_txt.params["m3/stride"]
    if m3 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m3_out_channel = model_info_txt.params["m3/out_channels"]
        m3_kernel = model_info_txt.params["m3/kernel_size"]
        m3_activation = model_info_txt.params["m3/activation"]
        m3_args = [m3_out_channel, m3_kernel, m3_stride, None, 1, m3_activation]
    elif m3 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m3_out_channel = model_info_txt.params["m3/out_channels"]
        m3_kernel = model_info_txt.params["m3/kernel_size"]
        m3_activation = model_info_txt.params["m3/activation"]
        m3_args = [m3_out_channel, m3_kernel, m3_stride, None, m3_activation]
    elif m3 == "InvertedResidualv2":
        m3_c = model_info_txt.params["m3/v2_c"]
        m3_t = model_info_txt.params["m3/v2_t"]
        m3_args = [m3_c, m3_t, m3_stride]
    elif m3 == "InvertedResidualv3":
        m3_kernel = model_info_txt.params["m3/kernel_size"]
        m3_t = model_info_txt.params["m3/v3_t"]
        m3_c = model_info_txt.params["m3/v3_c"]
        m3_se = model_info_txt.params["m3/v3_se"]
        m3_hs = model_info_txt.params["m3/v3_hs"]
        m3_args = [m3_kernel, m3_t, m3_c, m3_se, m3_hs, m3_stride]
    if not m3 == "Pass":
        if m3_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m3_repeat, m3, m3_args])

    # Module 4
    m4 = model_info_txt.params["m4"]
    m4_args = []
    m4_repeat = model_info_txt.params["m4/repeat"]
    m4_stride = model_info_txt.params["m4/stride"]
    # force stride m4
    if n_stride == 1:
        m4_stride = 2
    if m4 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m4_out_channel = model_info_txt.params["m4/out_channels"]
        m4_kernel = model_info_txt.params["m4/kernel_size"]
        m4_activation = model_info_txt.params["m4/activation"]
        m4_args = [m4_out_channel, m4_kernel, m4_stride, None, 1, m4_activation]
    elif m4 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m4_out_channel = model_info_txt.params["m4/out_channels"]
        m4_kernel = model_info_txt.params["m4/kernel_size"]
        m4_activation = model_info_txt.params["m4/activation"]
        m4_args = [m4_out_channel, m4_kernel, m4_stride, None, m4_activation]
    elif m4 == "InvertedResidualv2":
        m4_c = model_info_txt.params["m4/v2_c"]
        m4_t = model_info_txt.params["m4/v2_t"]
        m4_args = [m4_c, m4_t, m4_stride]
    elif m4 == "InvertedResidualv3":
        m4_kernel = model_info_txt.params["m4/kernel_size"]
        m4_t = model_info_txt.params["m4/v3_t"]
        m4_c = model_info_txt.params["m4/v3_c"]
        m4_se = model_info_txt.params["m4/v3_se"]
        m4_hs = model_info_txt.params["m4/v3_hs"]
        m4_args = [m4_kernel, m4_t, m4_c, m4_se, m4_hs, m4_stride]
    if not m4 == "Pass":
        if m4_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m4_repeat, m4, m4_args])

    # Module 5
    m5 = model_info_txt.params["m5"]
    m5_args = []
    m5_repeat = model_info_txt.params["m5/repeat"]
    m5_stride = 1
    if m5 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m5_out_channel = model_info_txt.params["m5/out_channels"]
        m5_kernel = model_info_txt.params["m5/kernel_size"]
        m5_activation = model_info_txt.params["m5/activation"]
        m5_stride = model_info_txt.params["m5/stride"]
        m5_args = [m5_out_channel, m5_kernel, m5_stride, None, 1, m5_activation]
    elif m5 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m5_out_channel = model_info_txt.params["m5/out_channels"]
        m5_kernel = model_info_txt.params["m5/kernel_size"]
        m5_activation = model_info_txt.params["m5/activation"]
        m5_stride = model_info_txt.params["m5/stride"]
        m5_args = [m5_out_channel, m5_kernel, m5_stride, None, m5_activation]
    elif m5 == "InvertedResidualv2":
        m5_c = model_info_txt.params["m5/v2_c"]
        m5_t = model_info_txt.params["m5/v2_t"]
        m5_stride = model_info_txt.params["m5/stride"]
        m5_args = [m5_c, m5_t, m5_stride]
    elif m5 == "InvertedResidualv3":
        m5_kernel = model_info_txt.params["m5/kernel_size"]
        m5_t = model_info_txt.params["m5/v3_t"]
        m5_c = model_info_txt.params["m5/v3_c"]
        m5_se = model_info_txt.params["m5/v3_se"]
        m5_hs = model_info_txt.params["m5/v3_hs"]
        m5_stride = model_info_txt.params["m5/stride"]
        m5_args = [m5_kernel, m5_t, m5_c, m5_se, m5_hs, m5_stride]
    if not m5 == "Pass":
        if m5_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m5_repeat, m5, m5_args])

    # Module 6
    m6 = model_info_txt.params["m6"]
    m6_args = []
    m6_repeat = model_info_txt.params["m6/repeat"]
    m6_stride = model_info_txt.params["m6/stride"]
    # force stride m6
    if n_stride == 2:
        m4_stride = 2
    if m6 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m6_out_channel = model_info_txt.params["m6/out_channels"]
        m6_kernel = model_info_txt.params["m6/kernel_size"]
        m6_activation = model_info_txt.params["m6/activation"]
        m6_args = [m6_out_channel, m6_kernel, m6_stride, None, 1, m6_activation]
    elif m6 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m6_out_channel = model_info_txt.params["m6/out_channels"]
        m6_kernel = model_info_txt.params["m6/kernel_size"]
        m6_activation = model_info_txt.params["m6/activation"]
        m6_args = [m6_out_channel, m6_kernel, m6_stride, None, m6_activation]
    elif m6 == "InvertedResidualv2":
        m6_c = model_info_txt.params["m6/v2_c"]
        m6_t = model_info_txt.params["m6/v2_t"]
        m6_args = [m6_c, m6_t, m6_stride]
    elif m6 == "InvertedResidualv3":
        m6_kernel = model_info_txt.params["m6/kernel_size"]
        m6_t = model_info_txt.params["m6/v3_t"]
        m6_c = model_info_txt.params["m6/v3_c"]
        m6_se = model_info_txt.params["m6/v3_se"]
        m6_hs = model_info_txt.params["m6/v3_hs"]
        m6_args = [m6_kernel, m6_t, m6_c, m6_se, m6_hs, m6_stride]
    if not m6 == "Pass":
        if m6_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m6_repeat, m6, m6_args])

    # Module 7
    m7 = model_info_txt.params["m7"]
    m7_args = []
    m7_repeat = model_info_txt.params["m7/repeat"]
    m7_stride = model_info_txt.params["m7/stride"]
    if m7 == "Conv":
        # Conv args: [out_channel, kernel_size, stride, padding, groups, activation]
        m7_out_channel = model_info_txt.params["m7/out_channels"]
        m7_kernel = model_info_txt.params["m7/kernel_size"]
        m7_activation = model_info_txt.params["m7/activation"]
        m7_args = [m7_out_channel, m7_kernel, m7_stride, None, 1, m7_activation]
    elif m7 == "DWConv":
        # DWConv args: [out_channel, kernel_size, stride, padding_size, activation]
        m7_out_channel = model_info_txt.params["m7/out_channels"]
        m7_kernel = model_info_txt.params["m7/kernel_size"]
        m7_activation = model_info_txt.params["m7/activation"]
        m7_args = [m7_out_channel, m7_kernel, m7_stride, None, m7_activation]
    elif m7 == "InvertedResidualv2":
        m7_c = model_info_txt.params["m7/v2_c"]
        m7_t = model_info_txt.params["m7/v2_t"]
        m7_args = [m7_c, m7_t, m7_stride]
    elif m7 == "InvertedResidualv3":
        m7_kernel = model_info_txt.params["m7/kernel_size"]
        m7_t = model_info_txt.params["m7/v3_t"]
        m7_c = model_info_txt.params["m7/v3_c"]
        m7_se = model_info_txt.params["m7/v3_se"]
        m7_hs = model_info_txt.params["m7/v3_hs"]
        m7_args = [m7_kernel, m7_t, m7_c, m7_se, m7_hs, m7_stride]
    if not m7 == "Pass":
        if m7_stride == 2:
            n_stride += 1
            if n_stride >= MAX_NUM_STRIDE:
                UPPER_STRIDE = 1
        model.append([m7_repeat, m7, m7_args])
    
    # last layer
    last_dim = model_info_txt.params["last_dim"]
    # We can setup fixed structure as well
    model.append([1, "Conv", [last_dim, 1, 1]])
    model.append([1, "GlobalAvgPool", []])
    model.append([1, "FixedConv", [6, 1, 1, None, 1, None]])

    module_info = {}
    module_info["m1"] = {"type": m1, "repeat": m1_repeat, "stride": m1_stride}
    module_info["m2"] = {"type": m2, "repeat": m2_repeat, "stride": m2_stride}
    module_info["m3"] = {"type": m3, "repeat": m3_repeat, "stride": m3_stride}
    module_info["m4"] = {"type": m4, "repeat": m4_repeat, "stride": m4_stride}
    module_info["m5"] = {"type": m5, "repeat": m5_repeat, "stride": m5_stride}
    module_info["m6"] = {"type": m6, "repeat": m6_repeat, "stride": m6_stride}
    module_info["m7"] = {"type": m7, "repeat": m7_repeat, "stride": m7_stride}

    return model, module_info
```

```python
model_config = {}
model_config["input_channel"] = 3
# img_size = model_info_txt.params[("img_size", [32, 64, 128])
img_size = 32
model_config["INPUT_SIZE"] = [img_size, img_size]
model_config["depth_multiple"] = model_info_txt.params['depth_multiple']
model_config["width_multiple"] = model_info_txt.params['width_multiple']
model_config["backbone"], module_info = search_model()
model_config
```
