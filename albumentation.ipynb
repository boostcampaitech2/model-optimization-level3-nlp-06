```python
import albumentations as A
import matplotlib.pyplot as plt
import cv2
```

```python
def visualize(image):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(image)
```

```python
image = cv2.imread('../data/train/Metal/00opbeliportgfoddnbjgklx.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
visualize(image)
```

```python
from albumentations.pytorch import ToTensorV2
def simple_augment_train(
    dataset: str = "CIFAR10", img_size: float = 32
) -> A.Compose:
    return A.Compose(
                [
                    A.LongestMaxSize(int(img_size * 1.2)),
                    A.PadIfNeeded(int(img_size * 1.2), int(img_size * 1.2), border_mode=0),
                    A.RandomCrop(img_size, img_size),
                    A.HorizontalFlip(p=0.5),
                    A.Normalize(    DATASET_NORMALIZE_INFO[dataset]["MEAN"],
                                    DATASET_NORMALIZE_INFO[dataset]["STD"],
                    ),
                    ToTensorV2(),
                    
                ])
```

```python
transform = simple_augment_train('a',224)
aug_image = transform(image=image)['image']
visualize(aug_image)
```
