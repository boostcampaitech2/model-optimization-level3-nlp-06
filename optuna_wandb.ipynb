```python
import optuna
from typing import Tuple
from optuna.integration import WeightsAndBiasesCallback
```

```python

def objective(trial: optuna.trial.Trial) -> Tuple[float, float, float]:
    x = trial.suggest_float('x',-10,10)
    return (x - 2) ** 2

wandbc = WeightsAndBiasesCallback(metric_name='y')
study = optuna.create_study(study_name='xy')
study.optimize(objective, n_trials=500, callbacks=[wandbc])
```
